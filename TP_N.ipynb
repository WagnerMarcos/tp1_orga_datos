```python
#Cantidad de posibles segundos usos 
cantSegundo = 10
```

```python
dfSegundoUso = df[df["has_secondary_use"] == 1].iloc[:,(cantSegundo*-1):]
prueba = dfSegundoUso.loc[(dfSegundoUso.sum(axis=1) != 1)]
prueba.head(5)
```

```python
prueba[prueba['has_secondary_use_other'] == 0]
```

Como llego a un dataframe vacio los unicos establecimientos que tienen mas de un segundo uso, es porque tienen "otro" uso que no es ninguno de los categorizados, osea que 9 columnas puedo resumirlas a una sola columna que sea de tipo categoria

```python
dfAcotado = df.iloc[:,(cantSegundo*-1):-1]
```

```python
dfSegundoUso = dfAcotado.dot(dfAcotado.columns).to_frame('segundo_uso')
dfSegundoUso['segundo_uso'] = dfSegundoUso['segundo_uso'].str.replace("has_secondary_use_","") # Borro la info inecesaria del string
dfSegundoUso['segundo_uso'] = dfSegundoUso['segundo_uso'].str.capitalize() # Hago la primera letra mayuscula
dfSegundoUso = dfSegundoUso.astype({'segundo_uso':'category'})
dfSegundoUso.value_counts()
```

```python
df = df.merge(dfSegundoUso, left_index=True, right_index=True)
```

```python
df.drop(columns = ['has_secondary_use_agriculture', 'has_secondary_use_hotel', 'has_secondary_use_rental', 'has_secondary_use_institution', 'has_secondary_use_school', 
                   'has_secondary_use_industry', 'has_secondary_use_health_post',  'has_secondary_use_gov_office', 'has_secondary_use_police_station'], inplace=True)
```

```python
df
```

```python
dfHasSecondary = df[df['has_secondary_use'] == 1].copy()
dfHasSecondary['segundo_uso'] = dfHasSecondary['segundo_uso'].replace({'':'Other', 
                                                                       'Health_post':'Health Post', 
                                                                       'Gov_office':'Gov Office',
                                                                       'Police_station':'Police Station'})
dfHasSecondary['segundo_uso'].value_counts()
```

```python
bar1 = dfHasSecondary.groupby('segundo_uso').agg({"damage_grade":"mean"}).sort_values(by='damage_grade')
bar1
```

```python
dfSuperSt = df.loc[:,['damage_grade','has_superstructure_adobe_mud',
       'has_superstructure_mud_mortar_stone', 'has_superstructure_stone_flag',
       'has_superstructure_cement_mortar_stone',
       'has_superstructure_mud_mortar_brick',
       'has_superstructure_cement_mortar_brick', 'has_superstructure_timber',
       'has_superstructure_bamboo', 'has_superstructure_rc_non_engineered',
       'has_superstructure_rc_engineered', 'has_superstructure_other']]

dfSuperSt['damage_grade'] = df['damage_grade'].apply(lambda x: d[x]) # d Es la lista de Alto Medio Bajo
```

```python
dfSuperStMean = dfSuperSt.groupby('damage_grade').agg('mean')
dfSuperStMean = dfSuperStMean.T
dfSuperStMean
```

```python
serieAPositiva = (dfSuperStMean['Alto']*100).rename('Positivo')
serieANegativa = (dfSuperStMean['Alto']*100-100).rename('Negativo')
miniA = pd.concat([serieAPositiva,  serieANegativa], axis=1).sort_values(by=['Positivo'])
miniA
```

```python
v = miniA.plot.barh( stacked=True, fontsize=12, color=["lightGreen","IndianRed"],\
              xlim=(-100.0, 100.0),width=0.5,figsize=(6, 7))
v.legend().remove()
v.set_title('T', fontsize=18);
v.set_xlabel('X de verdad', fontsize=15);
v.set_ylabel('Y', fontsize=15);
v.grid(axis='both')
```

```python
serieAPositiva = (dfSuperStMean['Medio']*100).rename('Positivo')
serieANegativa = (dfSuperStMean['Medio']*100-100).rename('Negativo')
miniM = pd.concat([serieAPositiva,  serieANegativa], axis=1).sort_values(by=['Positivo'])
miniM
```

```python
v = miniM.plot.barh( stacked=True, fontsize=12, color=["lightGreen","IndianRed"],\
              xlim=(-100.0, 100.0),width=0.5,figsize=(6, 7))
v.legend().remove()
v.set_title('T', fontsize=18);
v.set_xlabel('X de verdad', fontsize=15);
v.set_ylabel('Y', fontsize=15);
v.grid(axis='both')
```
